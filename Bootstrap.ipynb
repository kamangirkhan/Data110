```python
import pandas as pd
stocks = pd.read_csv("Stocks.csv")
print(stocks.head())
size = len(stocks)
print(size)
```

```python
import numpy as np
import matplotlib.pyplot as plt
resample = 1000
boot_means= []
for i in range(resample):
    sample = stocks.sample(n=size, replace=True)
    mean_return = sample["Shares"].mean()
    boot_means.append(mean_return)
ci_lower = np.percentile(boot_means, 2.5)
ci_upper = np.percentile(boot_means, 97.5)
print(f"Bootstrap Mean Return: {np.mean(boot_means):}")
```

```python
plt.hist(boot_means, bins=30, color='orange', edgecolor='black')
plt.title("Bootstrap Distribution of Mean return")
plt.xlabel("Mean Return")
plt.ylabel("Frequency")
plt.grid(True, axis = 'y', linestyle = '--')
plt.show()
```
