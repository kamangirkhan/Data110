```python
from googlesearch import search

results = []

# Original tutorial searches for books
items = [
    'free Java books',
    'free Python books',
    'Java IDE',
    'free computer books'
]

# Enhancement: search for "free data science courses"
items.append('free data science courses')

for query in items:
    print(f"Searching: {query}")
    for link in search(query, tld="com", num=7, stop=5, pause=4):
        results.append(query + "," + link)

print("\nDone Crawling.")
```

```python
with open("results.txt", "w") as f:
    for line in results:
        f.write(line + "\n")

print("Search results saved to results.txt")
```

```python
index = {}

with open("results.txt", "r") as f:
    for line in f:
        category, url = line.strip().split(",", 1)

        if category not in index:
            index[category] = [url]
        else:
            index[category].append(url)

index
```

```python
while True:
    term = input("\nWhat topic are you looking for? (type 'exit' to quit) ")

    if term.lower() == "exit":
        break

    found = False
    for key in index.keys():
        if term.lower() in key.lower():
            print(f"\nResults for '{key}':")
            for link in index[key]:
                print(" â€¢", link)
            found = True

    if not found:
        print("No results found.")
```
