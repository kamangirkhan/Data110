### 2. Basic Line Plot
Create a plot of the quadratic function $f(x) = x^2 - 2x + 1$ over the range $[-2, 4]$

```python
def f(x): return x**2 -2*x+1
x = np.linspace(-2, 4, 400)
y = f(x)
plt.plot(x, y, label=r'$f(x) = x^2 - 2x + 1$', color='b')
plt.xlabel('X')
plt.ylabel('$f(x) = x^2 - 2x + 1$')
plt.legend()
plt.show()
```

### 3. Multiple Lines
- Create a plot with the following three functions on the same graph:
  - $f(x) = x^3$
  - $g(x) = e^x$
  - $h(x) = \log(x)$ (for $x > 0$)

```python
def f(x): return x**3
def g(x): return np.exp(x)
def h(x): return np.log(x)

x = np.linspace (-2, 4, 400)
x_log = np.linspace(0.111, 4, 400)

y_f = f(x)
y_g = g(x)
y_h = h(x_log)

# Create the plot
plt.figure(figsize=(8, 6))

plt.plot(x, y_f, label=r'$f(x) = x^3$', color='r')
plt.plot(x, y_g, label=r'$g(x) = e^x$', color='g')
plt.plot(x_log, y_h, label=r'$h(x) = \log(x)$', color='y')

plt.xlabel('x')
plt.ylabel('Function values')
plt.title('Plot of $f(x) = x^3$, $g(x) = e^x$, and $h(x) = \log(x)$')

plt.legend()

plt.show()
```

### 4. Styling Lines
- Create a plot with the following functions, each with a different color and line style:
  - $f(x) = x^2$
  - $g(x) = x^3$

### 5. Axes Limits and Scaling
- Plot the function $f(x) = \frac{1}{x^2 + 1}$ over the range $[-10, 10]$
- Set appropriate custom x and y limits to showcase the function's behavior
- Create another plot of the same function with reversed x and y axes
- Demonstrate the use of `plt.axis('tight')` and `plt.axis('equal')`

```python
def f(x):  return 1 / (x**2 + 1)

x = np.linspace(-10, 10, 400)
y = f(x)

plt.figure(figsize=(8, 5))
plt.plot(y, x, label='Reversed Axes', color='r', linestyle='--')

plt.xlim(0, 1.2)
plt.ylim(10, -10)

plt.xlabel('f(x)')
plt.ylabel('x')
plt.title('Plot with Reversed Axes')
plt.legend()

plt.axis('tight')

plt.show()

plt.figure(figsize=(8, 5))
plt.plot(x, y, color='g', label='Equal Axis Scale')

plt.axis('equal')

plt.xlabel('x')
plt.ylabel('f(x)')
plt.title("Plot with 'equal' Aspect Ratio")
plt.legend()

plt.show()
```

### 6. Labels and Legends
- Create a plot of $f(x) = \sqrt{x}$ and $g(x) = x^{\frac{1}{3}}$ for $x \geq 0$
- Add a proper title, x-label, y-label, and legend

```python
def f(x): return np.sqrt(x)
def g(x): return np.cbrt(x)

x = np.linspace(0, 10 , 400)
y_f = f(x)
y_g = g(x)

plt.figure(figsize=(8, 5))
plt.plot(x, y_f, label=r'$f(x)= \sqrt{x}$', color= 'r', linestyle= '--')
plt.plot(x, y_g, label=r'$g(x) = x^{\frac{1}{3}}$', color= 'k', linestyle= ':')

plt.xlabel('X')
plt.ylabel('Functions')
plt.title('Plot of $f(x) = \sqrt{x}$ and $g(x) = x^{1/3}$')
plt.legend()

plt.show()
```

### 8. Customizing Scatter Plots
- Generate data for the function $f(x) = x \sin(x)$ over the range $[0, 4\pi]$
- Create a scatter plot of this data with a line connecting the points
- Customize the plot by changing the marker size, line width, marker face color, marker edge color, and marker edge width

```python
def f(x): return x*np.sin(x)
x = np.linspace(0, 4*np.pi, 100)
y = f(x)

plt.figure(figsize=(8, 5))
plt.plot(x, y, label=r'$line$', linestyle='-', color='b', linewidth=1)
plt.scatter(x, y, s=25, label='$points$', color='w', edgecolor='r', linewidth=1)

plt.xlabel('X')
plt.ylabel('Y')
plt.title('$x \sin(x)$')
plt.legend()
plt.show()
```

### 9. Combined Plot
- Plot the function $f(x) = x \cos(x)$ over the range $[-2\pi, 2\pi]$
- Add scatter points for x values at every $\frac{\pi}{4}$ interval
- Use different colors and styles for the line and scatter components

```python
def f(x): return x*np.cos(x)
x = np.linspace(-2*np.pi, 2*np.pi, 100)
y = f(x)

plt.figure(figsize=(10, 8))
x_scatter = np.arange(-2*np.pi, 2*np.pi, np.pi/4)
y_scatter = x_scatter * np.cos(x_scatter)

plt.plot(x, y, label=r'$line$', linestyle='--', color='b', linewidth=2)
plt.scatter(x_scatter, y_scatter, s=50, label="points" , color='red', edgecolor='black', linewidth=1.5)

plt.xlabel('X')
plt.ylabel('Y')
plt.title('$x \cos(x)$')
plt.legend()

plt.show()
```

### 10. Custom Plot with Alternative Style
- Choose two or more functions from the following list:
  - $f(x) = \tan(x)$
  - $g(x) = x e^{-x^2}$
  - $h(x) = \frac{\sin(x)}{x}$  (with appropriate handling for $x=0$)
  - $i(x) = \arctan(x)$
  - $j(x) = \frac{1}{1 + e^{-x}}$  (logistic function)
- Create a custom plot that combines at least 4 concepts learned (e.g., multiple functions, scatter points, custom colors, line styles, limits, labels, and legend)
- Use the alternative style you chose in Task 1 for this plot
- Add comments explaining your choices, the significance of the functions you chose, and how the alternative style affects the plot's appearance


```python
import numpy as np
import matplotlib.pyplot as plt

# Set an alternative style
plt.style.use('ggplot')  # Alternative style for a clear and professional look

# Generate x values
x = np.linspace(-5, 5, 400)  # Covers a reasonable range, avoiding discontinuities in tan(x)

# Define functions
g_x = x * np.exp(-x**2)          # g(x) = x * e^(-x^2)
j_x = 1 / (1 + np.exp(-x))       # j(x) = 1 / (1 + e^(-x)) (logistic function)

# Define scatter points at key locations
#x_scatter = np.array([-2, -1, 0, 1, 2])
x_scatter = np.arange(-5, 5, 1)
y_scatter_g = x_scatter * np.exp(-x_scatter**2)
y_scatter_j = 1 / (1 + np.exp(-x_scatter))

# Create the plot
plt.figure(figsize=(12, 7))

# Plot each function with different styles
plt.plot(x, g_x, linestyle='--', color='b', linewidth=2, label=r'$g(x) = x e^{-x^2}$')
plt.plot(x, j_x, linestyle='-', color='orange', linewidth=2, label=r'$j(x) = \frac{1}{1+e^{-x}}$')

# Scatter points for selected functions
plt.scatter(x_scatter, y_scatter_g, color='blue', edgecolor='black', s=50, label="g(x) Key Points")
plt.scatter(x_scatter, y_scatter_j, color='orange', edgecolor='black', s=50, label="j(x) Key Points")

# Customize axes and limits
plt.axhline(0, color='gray', linestyle='--', linewidth=1)  # Reference horizontal line
plt.axvline(0, color='gray', linestyle='--', linewidth=1)  # Reference vertical line
plt.xlim(-5, 5)
plt.ylim(-2, 2)  # Adjusted to keep tan(x) within visible range

# Add labels, title, and legend
plt.xlabel("x-axis")
plt.ylabel("Function values")
plt.title(r'Plot of $g(x) = \sqrt{x}$ and $j(x) = \frac{1}{1+e^{-x}}$')
plt.legend()

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Set an alternative style
plt.style.use('ggplot')  # Alternative style for a clear and professional look

# Generate x values
x = np.linspace(-2*np.pi, 2*np.pi, 400)  # Covers a reasonable range, avoiding discontinuities in tan(x)

# Define functions
f_x = np.tan(x)                  # f(x) = tan(x)
h_x = np.where(x == 0, 1, np.sin(x) / x)  # h(x) = sin(x)/x with h(0) = 1
i_x = np.arctan(x)               # i(x) = arctan(x)

# Define scatter points at key locations
#x_scatter = np.array([-2, -1, 0, 1, 2])
x_scatter = np.arange(-2*np.pi, 2*np.pi, np.pi/6)
y_scatter_f = np.tan(x_scatter)
y_scatter_h = np.sinc(x_scatter / np.pi)  # np.sinc() is normalized, so adjust accordingly
y_scatter_i = np.arctan(x_scatter)

# Create the plot
plt.figure(figsize=(12, 7))

# Plot each function with different styles
plt.plot(x, f_x, linestyle='-', color='purple', linewidth=2, label=r'$f(x) = \tan(x)$')
plt.plot(x, h_x, linestyle='-.', color='r', linewidth=2, label=r'$h(x) = \frac{\sin(x)}{x}$')
plt.plot(x, i_x, linestyle=':', color='g', linewidth=2, label=r'$i(x) = \arctan(x)$')

# Scatter points for selected functions
plt.scatter(x_scatter, y_scatter_f, color='blue', edgecolor='black', s=50, label="f(x) Key Points")
plt.scatter(x_scatter, y_scatter_h, color='red', edgecolor='black', s=50, label="h(x) Key Points")
plt.scatter(x_scatter, y_scatter_i, color='green', edgecolor='black', s=50, label="i(x) Key Points")

# Customize axes and limits
plt.axhline(0, color='gray', linestyle='--', linewidth=1)  # Reference horizontal line
plt.axvline(0, color='gray', linestyle='--', linewidth=1)  # Reference vertical line
plt.xlim(-2*np.pi, 2*np.pi)
plt.ylim(-20, 20)  # Adjusted to keep tan(x) within visible range

# Add labels, title, and legend
plt.xlabel("x-axis")
plt.ylabel("Function values")
plt.title(r'Plot of $f(x) = \tan(x)$, $h(x) = \frac{\sin(x)}{x}$ and $i(x) = \arctan(x)$')
plt.legend()

# Show the plot
plt.show()
```
