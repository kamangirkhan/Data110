###  Anscombe's Quartet Dataset

This dataset is known as **Anscombe's Quartet**, created by statistician Francis Anscombe to illustrate the importance of visualizing data. Despite having nearly identical statistical properties (e.g., mean, variance, correlation, and linear regression), each dataset tells a very different story when graphed.

- **x**: The independent variable, common across three datasets.
- **y1, y2, y3**: Three different dependent variables associated with the same `x` values.
- **x4, y4**: A special case where most of the `x` values are identical, with one outlier.

#### Anscombe's Quartet:

```python
# Anscombe's Quartet:

x = [10, 8, 13, 9, 11, 14, 6, 4, 12, 7, 5]
y1 = [8.04, 6.95, 7.58, 8.81, 8.33, 9.96, 7.24, 4.26, 10.84, 4.82, 5.68]
y2 = [9.14, 8.14, 8.74, 8.77, 9.26, 8.10, 6.13, 3.10, 9.13, 7.26, 4.74]
y3 = [7.46, 6.77, 12.74, 7.11, 7.81, 8.84, 6.08, 5.39, 8.15, 6.42, 5.73]
x4 = [8, 8, 8, 8, 8, 8, 8, 19, 8, 8, 8]
y4 = [6.58, 5.76, 7.71, 8.84, 8.47, 7.04, 5.25, 12.50, 5.56, 7.91, 6.89]
```

```python
# Dataset 1: A typical linear relationship.

plt.scatter(x, y1)
plt.xlabel('x')
plt.ylabel('y1')
plt.title('(x, y1) Data Points')
plt.text(max(x) -4, min(y1), s="A typical linear relationship")
plt.show()
```

```python
# Calculate the slope (m) and intercept (b) of the line using np.polyfit
m, b = np.polyfit(x, y1, 1)
regression_line = m * np.array(x) + b
plt.scatter(x, y1)
plt.plot(x, regression_line,color='blue')
plt.xlabel('x')
plt.ylabel('y1')
plt.title('(x, y1) Data Points and Regression Line')
plt.text(max(x) -7, min(y1), s="A typical linear relationship with regression line")
plt.show()
```

```python
# Dataset 2: A non-linear relationship

plt.scatter(x, y2)
plt.xlabel('x')
plt.ylabel('y2')
plt.title('(x, y2) Data Points')
plt.text(max(x) -3.5, min(y1) -1.25, s="A non-linear relationship")
plt.show()
```

```python
# Calculate the slope (m) and intercept (b) of the line using np.polyfit
m, b = np.polyfit(x, y2, 1)
regression_line = m * np.array(x) + b
plt.scatter(x, y2)
plt.plot(x, regression_line,color='blue')
plt.xlabel('x')
plt.ylabel('y2')
plt.title('(x, y2) Data Points and Regression Line')
plt.text(max(x) -7, min(y1) -1, s="A non-linear relationship with regression line")
plt.show()
```

```python
# Dataset 3: A dataset with an outlier

plt.scatter(x, y3)
plt.xlabel('x')
plt.ylabel('y3')
plt.title('(x, y3) Data Points')
plt.text(max(x) -4, min(y1) +1, s="A dataset with an outlier")
plt.show()
```

```python
# Calculate the slope (m) and intercept (b) of the line using np.polyfit
m, b = np.polyfit(x, y3, 1)
regression_line = m * np.array(x) + b

plt.scatter(x, y3)
plt.plot(x, regression_line,color='blue')
plt.xlabel('x')
plt.ylabel('y3')
plt.title('(x, y3) Data Points and Regression Line')
plt.text(max(x) -7, min(y1) +1, s="A dataset with an outlier with regression line")
plt.show()
```

```python
# Dataset 4: A vertical pattern caused by a single influential point

plt.scatter(x4, y4)
plt.xlabel('x4')
plt.ylabel('y4')
plt.title('(x4, y4) Data Points')
plt.text(max(x) -4, min(y1) +1, s="A vertical pattern caused by a single influential point")
plt.show()
```

```python
# Calculate the slope (m) and intercept (b) of the line using np.polyfit
m, b = np.polyfit(x4, y4, 1)
regression_line = m * np.array(x) + b

plt.scatter(x4, y4)
plt.plot(x4, regression_line,color='blue')
plt.xlabel('x4')
plt.ylabel('y4')
plt.title('(x4, y4) Data Points and Regression Line')
plt.text(min(x) +4, max(y1) +1, s="A vertical pattern caused by a single influential point\n"
"with regression line")
plt.show()
```
