1.	Using RegEx, split a text into a list of words

    •	txt = 'it is raining outside. we could use the rain.'
  
    o	Output: ['it', 'is', 'raining', 'outside', 'we', 'could', 'use', 'the', 'rain']
  
    o	You are here to practice RegEx

```python
import re

txt = 'it is raining outside. we could use the rain.'

# Use re.findall() to extract only the words
words = re.findall(r'\b\w+\b', txt)

print(words)
```

2.	Using RegEx, Implement a function to determine if an input string is in a proper English sentence structure.
  
•	def is_sentence(text)

    text is the input string
    
    return True or False
    
    For this assignment, a sentence begins with a capital letter and ends with a period (full stop)
    
•	Expand on this definition if you like

    is_sentence(“This is a test.”)   #returns True
    
    is_sentence(“hey you!) # returns False


```python
import re

def is_sentence(text):
    # A proper sentence starts with a capital letter,
    # followed by any characters, and ends with a period.
    pattern = r'^[A-Z][^.?!]*\.$'
    return bool(re.match(pattern, text))

# ✅ Test cases
print(is_sentence("This is a test."))   # True
print(is_sentence("hey you!"))          # False
print(is_sentence("Hello there"))       # False
print(is_sentence("Welcome!"))          # False
```
