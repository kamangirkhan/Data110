```python
import pandas as pd
data = {
    'Offense': ['Aggravated', 'Burglary', 'Larceny', 'Motor Vehicle Theft' ,'Rape (revised definition)', 'Robbery'],
    'Offense per 100,000 People (2023)': [264.1, 250.7, 1347.2, 318.7, 38.0, 66.5],
    'Change in Offenses per 100,000 People (2022-2023)': [-3.3, -8.1, -4.9, +12.0, -9.8, -0.8]
}
df=pd.DataFrame(data)
df
```

## Task 1: Basic Bar Chart for Crime Rates in 2023

In this task, you will create a **bar chart** showing the offense rates per 100,000 people for each type of crime in 2023.

### Instructions:
- Create a bar chart for crime rates in 2023.
- Customize the bar colors and add gridlines.
- Label the axes and add a title.

```python
import matplotlib.pyplot as plt

colors = ['blue', 'red', 'green', 'orange', 'purple', 'brown']

plt.figure(figsize=(12, 8))
plt.bar(df['Offense'], df['Offense per 100,000 People (2023)'], color=colors, edgecolor='black')

plt.xlabel('Rate', fontsize=12)
plt.ylabel('Crime', fontsize=12)
plt.title('Crime Rates per 100,000 People in 2023', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

## Task 2: Change in Crime Rates (2022 vs. 2023) with Unique Colors

In this task, you will create a **horizontal bar chart** that visualizes the **percentage change** in offense rates from 2022 to 2023 for each crime type. Each bar should have a unique color for better distinction.

### Instructions:

   - Explain why seeing the **change** over time is more insightful than just seeing the total number of offenses for a single year.
   
2. **Improve the clarity of your bar graph**:
   - Create **multiple versions** of the bar chart for Task 2, each time making the chart more **clear and readable**.
   - Experiment with different color schemes, labels, and gridlines.
   - In each iteration, focus on improving the ease with which someone can quickly understand the trends in crime rates.

```python
import matplotlib.pyplot as plt

colors = ['blue', 'red', 'green', 'orange', 'purple', 'brown']

plt.figure(figsize=(12,8))
plt.barh(df['Offense'], df['Change in Offenses per 100,000 People (2022-2023)'], color=colors, edgecolor='black')

plt.xlabel('Change in Offense Rate per 100,000 People', fontsize=12)
plt.ylabel('Crimes', fontsize=12)
plt.title('Change in Offenses per 100,000 People (2022-2023)', fontsize=14)
plt.grid()

plt.show()
```

- Task 1s graph is useful for knowing how common each crime was in 2023.
- This graph is more useful for identifying trendswhich crimes are increasing or decreasing.
- Change over time helps decision-makers focus on the most urgent issues, not just the biggest categories.
